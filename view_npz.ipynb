```python
import numpy as np

path = "results/xai/umamba_mtl/fold_0/10008_1000008.npz"

d = np.load(path)
for k in d:
    arr = d[k]
    print(f"{k:30s}  shape={str(arr.shape):25s}  dtype={arr.dtype}  min={arr.min():.4g}  max={arr.max():.4g}")
```

```python
import matplotlib.pyplot as plt

key = "image"   # change to any key above
arr = d[key]

# squeeze to 3D if needed, then show middle slice
while arr.ndim > 3:
    arr = arr[0]

mid = arr.shape[0] // 2
plt.imshow(arr[mid], cmap="gray")
plt.title(f"{key}  slice {mid}/{arr.shape[0]}")
plt.colorbar()
plt.show()
```
